```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
%matplotlib inline
```

```python
f = open('Particle_Track.txt', 'r')
data = f.read().splitlines()
f.close()
float_data = []
for i in range(len(data)):
    float_data.append(float(data[i]))
```

```python
x = range(len(data))
x = np.array(x)
plt.plot(x,float_data,'-o')
plt.hlines(1,0,100)
plt.hlines(2,0,100)
plt.hlines(3,0,100)
```
