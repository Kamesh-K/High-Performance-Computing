```python
import seaborn as sns 
import pandas as pd 
import numpy as np
import time
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Temperature_Profile.txt')
```

```python
data.head(-10)
```

```python
for i in range(20):
    plt.clf()
    a = data.iloc[i,:]
    a = np.asarray(a)
    a = np.reshape(a,(8,8))
    sns.heatmap(a)
    time.sleep(1)
```
